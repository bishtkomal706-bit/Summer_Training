```python
df=df.drop_duplicates()
```

```python
# Create price category column
def price_category(price):
    if price <= 10000:
        return 'Budget'
    elif price <= 30000:
        return 'Mid'
    else:
        return 'Premium'

df['price_category'] = df['price'].apply(price_category)
print(df['price_category'].value_counts())
```

```python
# X and Y for Classifier
X_clf = df.drop(['price', 'price_category'], axis=1)
Y_clf = df['price_category']
```

```python
# Encoding
le2 = LabelEncoder()
X_clf['airline'] = le2.fit_transform(X_clf['airline'])
X_clf['source_city'] = le2.fit_transform(X_clf['source_city'])
X_clf['departure_time'] = le2.fit_transform(X_clf['departure_time'])
X_clf['stops'] = le2.fit_transform(X_clf['stops'])
X_clf['arrival_time'] = le2.fit_transform(X_clf['arrival_time'])
X_clf['destination_city'] = le2.fit_transform(X_clf['destination_city'])
X_clf['class'] = le2.fit_transform(X_clf['class'])
```

```python
# Split
X_train_c, X_test_c, Y_train_c, Y_test_c = train_test_split(X_clf, Y_clf, test_size=0.2, random_state=42)

print("Train size:", X_train_c.shape)
print("Test size:", X_test_c.shape)
```

```python
# Model
clf_model = DecisionTreeClassifier(max_depth=15, min_samples_split=20, min_samples_leaf=10, random_state=42)
clf_model.fit(X_train_c, Y_train_c)
y_pred_clf = clf_model.predict(X_test_c)
```

```python
# Evaluation
print("Train Score:", round(clf_model.score(X_train_c, Y_train_c), 2))
print("Test Score :", round(clf_model.score(X_test_c, Y_test_c), 2))
print("Accuracy   :", round(accuracy_score(Y_test_c, y_pred_clf), 2))
print()
print("Classification Report:")
print(classification_report(Y_test_c, y_pred_clf))
print()
print("Confusion Matrix:")
print(confusion_matrix(Y_test_c, y_pred_clf))
```
