```python
df=df.drop_duplicates()
```

```python
# Redo X and Y for Regressor
X_reg = df.drop(['price'], axis=1)  # keep class this time, whole column except price
Y_reg = df['price']                 # target / predicting price
```

```python
# Encode all categorical columns
le = LabelEncoder()
X_reg['airline'] = le.fit_transform(X_reg['airline'])
X_reg['source_city'] = le.fit_transform(X_reg['source_city'])
X_reg['departure_time'] = le.fit_transform(X_reg['departure_time'])
X_reg['stops'] = le.fit_transform(X_reg['stops'])
X_reg['arrival_time'] = le.fit_transform(X_reg['arrival_time'])
X_reg['destination_city'] = le.fit_transform(X_reg['destination_city'])
X_reg['class'] = le.fit_transform(X_reg['class'])
```

```python

# Train test split
X_train, X_test, Y_train, Y_test = train_test_split(X_reg, Y_reg, test_size=0.2, random_state=42)
```

```python
# Model
reg_model = DecisionTreeRegressor(max_depth=15, min_samples_split=20, min_samples_leaf=10, random_state=42)
reg_model.fit(X_train, Y_train)
y_pred_reg = reg_model.predict(X_test)
```

```python
print("Train Score:", round(reg_model.score(X_train, Y_train), 2))
print("Test Score :", round(reg_model.score(X_test, Y_test), 2))
print("R2 Score   :", round(r2_score(Y_test, y_pred_reg), 2))
```
