```python
import matplotlib.pyplot as plt                                     #pyplot--> for drawing graphs
import matplotlib.patches as patches                                #patches--> for drawing shapes like circle,arc,rectangle

fig, ax = plt.subplots(1,1,figsize=(5,5))                          # 1,1 -> 1 row and col only 1 plot, window size of 5 by 5 inches square shape
face = plt.Circle((0.5,0.5),0.4,color='yellow',ec='black',lw=2)   #big round face center at middle of 0.5,0.5 with radius 0.4, edge color , line width
ax.add_patch(face)                                                #circle is created but not visible yet .. so paste it on the plot --> add_patch

left_eye = plt.Circle((0.35,0.62),0.05,color='black')   
ax.add_patch(left_eye)

right_eye = plt.Circle((0.65,0.62),0.05,color='black')   
ax.add_patch(right_eye)

smile=patches.Arc((0.5,0.45),0.3,0.2,angle=0,theta1=200,theta2=340,color='black',lw=2)     #(centre of arc),width of arc,height,dont rotate the arc(angle=0),
ax.add_patch(smile)                                                                        #theta1-->where arc starts, theta2-->arc ends

ax.set_xlim(0,1)                                                #x axis goes from 0 to 1
ax.set_ylim(0,1)                                                #y axis gpes from 0 to 1
ax.set_title('Smiley Face')         
ax.axis('off')                                                  #hide the border box and numbers

plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
x=[1,2,3,4,5]
y=[10,20,30,40,50]
y1=[[10,20,30,40,50],
    [1,2,3,4,5]]
ax1= plt.subplot(331)
ax2= plt.subplot(332)
ax3= plt.subplot(333)
ax4= plt.subplot(334)
ax5= plt.subplot(335)
ax6= plt.subplot(336)
ax7= plt.subplot(337)
ax8= plt.subplot(338)
ax9= plt.subplot(339)
ax1.plot(x,color='y', linestyle = '--', marker='o')
ax1.set_title('Line Plot')
ax1.set_xlabel('Index')
ax1.set_ylabel('Value')

ax2.stackplot(x,y,color='c')
ax2.set_title('Area Plot')
ax2.set_xlabel('X')
ax2.set_ylabel('Y')

ax3.boxplot(x)
ax3.set_title('Box Plot')
ax3.set_xlabel('X')
ax3.set_ylabel('Y')

ax4.hist(x,color='r')
ax4.set_title('Histogram Plot')
ax4.set_xlabel('Bins')
ax4.set_ylabel('Frequency')

ax5.bar(x,y,color=['r','g','c','m','b'])
ax5.set_title('Bar Plot')
ax5.set_xlabel('Index')
ax5.set_ylabel('Value')

ax6.violinplot(x)
ax6.set_title('Violin Plot')
ax6.set_xlabel('Data')
ax6.set_ylabel('Value')

ax7.barh(x,y,color='g')
ax7.set_title('Horizontal Bar Plot')
ax7.set_xlabel('Index')
ax7.set_ylabel('Value')

ax8.scatter(x,y,color='m',marker='^') 
ax8.set_title('Scatter Plot')
ax8.set_xlabel('X')
ax8.set_ylabel('Data Value')

ax9.stackplot(x,y1)
ax9.set_title('Stack Area Plot')
ax9.set_xlabel('No. Of points')
ax9.set_ylabel('Value')
plt.subplots_adjust(hspace=1.5,wspace=1)

plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
x=[1,2,3,4,5]
y=[10,20,30,40,50]
y1=[[10,20,30,40,50],
    [1,2,3,4,5]]

fig,axes=plt.subplots(3,3, figsize=(8,8))
plt.subplots_adjust(hspace=0.6,wspace=0.5)
fig.suptitle("Different Types Of Matplotlib graphs",fontsize=14)
ax1= axes[0][0]
ax2= axes[0][1]
ax3= axes[0][2]
ax4= axes[1][0]
ax5= axes[1][1]
ax6= axes[1][2]
ax7= axes[2][0]
ax8= axes[2][1]
ax9= axes[2][2]
ax1.plot(x,color='y', linestyle = '--', marker='o')
ax1.set_title('Line Plot')
ax1.set_xlabel('Index')
ax1.set_ylabel('Value')

ax2.stackplot(x,y,color='c')
ax2.set_title('Area Plot')
ax2.set_xlabel('X')
ax2.set_ylabel('Y')

ax3.boxplot(x)
ax3.set_title('Box Plot')
ax3.set_xlabel('X')
ax3.set_ylabel('Y')

ax4.hist(x,color='r')
ax4.set_title('Histogram Plot')
ax4.set_xlabel('Bins')
ax4.set_ylabel('Frequency')

ax5.bar(x,y,color=['r','g','c','m','b'])
ax5.set_title('Bar Plot')
ax5.set_xlabel('Index')
ax5.set_ylabel('Value')

ax6.violinplot(x)
ax6.set_title('Violin Plot')
ax6.set_xlabel('Data')
ax6.set_ylabel('Value')

ax7.barh(x,y,color='g')
ax7.set_title('Horizontal Bar Plot')
ax7.set_xlabel('Index')
ax7.set_ylabel('Value')

ax8.scatter(x,y,color='m',marker='^')
ax8.set_title('Scatter Plot')
ax8.set_xlabel('X')
ax8.set_ylabel('Data Value')

ax9.stackplot(x,y1)
ax9.set_title('Stack Area Plot')
ax9.set_xlabel('No. Of points')
ax9.set_ylabel('Value')

plt.show()
```

# Pie Chart

```python
import matplotlib.pyplot as plt
sizes=[10,20,30,40]
labels=['a','b','c','d']

explode=(0,0,0,0)

ax1=plt.subplot()
ax1.pie(sizes,explode=explode,labels=labels,autopct='%9.2f%%')  #9 is any no representing whole no, 2 is total no of digit after decimal, f is floating, 
plt.show()                                                      # % is calculating %age itself
```

```python
#Create a piechart with 5 student marks & highlight/explode which is havig highest marks

import matplotlib.pyplot as plt
marks=[90,96,89,76,55]
student=['Karan','Komal','Priyanshu','Parth','Nisha']
explode=(0,0.1,0,0,0)
ax1 = plt.subplot()
ax1.pie(marks, explode = explode, labels = student, autopct='%1.2f%%')
plt.show()
```
