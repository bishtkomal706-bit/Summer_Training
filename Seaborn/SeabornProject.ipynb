```python
import pandas as pd
df = pd.read_csv(r"D:\DATA_SCIENCE\height-weight.csv")
df
```

```python
X = df[['Weight']]
Y = df['Height']
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,Y_train,Y_test = train_test_split(X,Y,test_size=0.20,random_state=42)
```

```python
#randomly choosen following weight is trained to model
X_train 
```

```python
Y_train                  #height i.e. ans of above random weight whose ans our model knows already
```

```python
X_test                           #these following data is choosen for testing the model whose ans model dont know
```

```python
Y_test                           #ans of above testing ques.. then we will compare the ans of above test ques with model's prediction
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score
regression = LinearRegression()
regression.fit(X_train , Y_train)
y_pred = regression.predict(X_test)
score = r2_score(Y_test , y_pred)
print(score)
```

```python
q1 = df['Height'].quantile(0.25)
q3 = df['Height'].quantile(0.75)
print(q1,q3)
IQR = q3-q1
print(IQR)

lower = q1 - 1.5 * IQR
upper = q3 + 1.5 * IQR
print(lower,upper)

outliers = df[(df['Height']<lower) | (df['Height']>upper)]
print(outliers)
```

```python
#Z-Score Method ( any value of z will be treated as outlier if it is less than -3 or greater than +3 )

a = df['Height'].mean()
b = df['Height'].std()
z = (df['Height'] - a) / b     # Z = (value(X)-Mean) / S.D.
z
```
