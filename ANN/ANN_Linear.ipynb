```python
x = data[['credit_score', 'age', 'tenure', 'products_number', 'credit_card', 'active_member', 'churn']]
y = data['estimated_salary']
```

```python
x_train ,x_test, y_train,y_test = train_test_split(x,y,test_size=0.3,random_state=123)
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense

linmodel = Sequential([
    Dense(64, activation='relu', input_dim=7),
    Dense(32, activation='relu'),
    Dense(1)  # Linear ke liye koi activation nahi
])

linmodel.compile(optimizer='adam', loss='mse', metrics=['mae'])

linmodel.fit(x_train, y_train, epochs=50, batch_size=10, validation_split=0.1)
```

```python
mse=linmodel.evaluate(x_test,y_test)
```

```python
print(f'Test Mean Squared Error:{mse}')
```
