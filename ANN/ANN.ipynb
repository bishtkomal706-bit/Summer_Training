```python
import numpy as np
import pandas as pd
import keras
import tensorflow as tf
from keras.models import Sequential
import keras.models as km
from keras import activations, initializers, regularizers, constraints
from keras.layers import Dense, Activation
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
```

```python
df = pd.read_csv(r"D:\winequality-red.csv")
df
```

```python
# dense-define layers
# hidden layer---unki value jb  uniform aane lge smjh lena vahi hidden layer h
# 1st and last ko chod ke bache huye bich ke sare hidden layer honge
```

```python
df['quality'].value_counts()
```

```python
x = df.drop(columns=['quality'])
y = df['quality']
```

```python
x_train ,x_test, y_train,y_test = train_test_split(x,y,test_size=0.3,random_state=123)
```

```python
x_train.shape
```

```python
model=Sequential()
model.add(Dense(units=11,kernel_initializer='uniform',activation='relu',input_dim=11))
model.add(Dense(units=6,kernel_initializer='uniform',activation='relu'))
model.add(Dense(units=4,kernel_initializer='uniform',activation='relu'))
model.add(Dense(units=1,kernel_initializer='uniform',activation='relu'))
```

```python
model.summary()
```

```python
model.compile(loss='mean_squared_error', optimizer='adam')
```

```python
print(x_train.shape)
```

```python
print(y_train.shape)
```

```python
model.fit(x_train,y_train,epochs=50,batch_size=32,validation_split=0.1)
```

```python
mse=model.evaluate(x_test,y_test)
```

```python
print(f'Test Mean Squared Error:{mse}')
```

```python
predictions = model.predict(x_test)
```
