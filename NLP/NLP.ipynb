```python
import nltk
nltk.download('wordnet')
```

```python
from nltk.stem import WordNetLemmatizer
```

```python
k = ["eat","eating","walk","walking"]
```

```python
#lemmatizer --> changes word into in its normal form eg; worked--> work
lem = WordNetLemmatizer()  
```

```python
for i in k:
    print(i,lem.lemmatize(i,pos='v'))
```
