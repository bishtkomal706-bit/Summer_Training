```python
# Create price category column
def price_category(price):
    if price <= 10000:
        return 'Budget'
    elif price <= 30000:
        return 'Mid'
    else:
        return 'Premium'

df['price_category'] = df['price'].apply(price_category)
print(df['price_category'].value_counts())
```

```python
X_clf = df.drop(['price', 'price_category'], axis=1)
Y_clf = df['price_category']
```

```python
le2 = LabelEncoder()
X_clf['airline'] = le2.fit_transform(X_clf['airline'])
X_clf['source_city'] = le2.fit_transform(X_clf['source_city'])
X_clf['departure_time'] = le2.fit_transform(X_clf['departure_time'])
X_clf['stops'] = le2.fit_transform(X_clf['stops'])
X_clf['arrival_time'] = le2.fit_transform(X_clf['arrival_time'])
X_clf['destination_city'] = le2.fit_transform(X_clf['destination_city'])
X_clf['class'] = le2.fit_transform(X_clf['class'])
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(X_clf, Y_clf, test_size=0.2, random_state=42)
```

```python
clf_model2 = RandomForestClassifier(max_depth=15, min_samples_split=20, min_samples_leaf=10, random_state=42)
clf_model2.fit(X_train, Y_train)
y_predict2 = clf_model2.predict(X_test)
```

```python
clf_model2.score(X_train,Y_train)
```

```python
clf_model2.score(X_test,Y_test)
```
